```python
import threading, time
def func1(epoch):
    time.sleep(0.1)
    print(epoch+",1", end="\n")
    
def func2(epoch):
    time.sleep(0.01)
    print(epoch+",2", end="\n")
    
def func3(epoch):
    time.sleep(0.01)
    print(epoch+",3", end="\n")
    
def func4(epoch):
    time.sleep(0.01)
    print(epoch+",4", end="\n")
    
def func5(epoch):
    time.sleep(0.01)
    print(epoch+",5", end="\n")
    
def func6(epoch):
    time.sleep(0.01)
    print(epoch+",6", end="\n")

def func7(epoch):
    time.sleep(0.1)
    print(epoch+",7", end="\n")
    
def func8(epoch):
    time.sleep(0.01)
    print(epoch+",8", end="\n")
    
def func9(epoch):
    time.sleep(0.01)
    print(epoch+",9", end="\n")
    
def func10(epoch):
    time.sleep(0.2)
    print(epoch+",10", end="\n")
    
def func11(epoch):
    time.sleep(0.8)
    print(epoch+",11", end="\n")
```

```python
listparallel = []
for p in range(11,21):
    epochs = p
    queue = [i for i in range(1,epochs+1)]
    calcul  = 11 + epochs - 1
    funq = [func11,func10,func9,func8,func7,func6,func5,func4,func3,func2,func1]
    tic = time.time()
    for i in range(calcul):
        threads=[]
        print("TIME "+str(i+1))
        if(i==0):
            threadsta = threading.Thread(target=func1,args=(str(queue[0]),))
            threads.append(threadsta)
        if(i==1):
            threadstb = threading.Thread(target=func2,args=(str(queue[0]),))
            threads.append(threadstb)
            threadsta = threading.Thread(target=func1,args=(str(queue[1]),))
            threads.append(threadsta)
        if(i==2):
            threadstc = threading.Thread(target=func3,args=(str(queue[0]),))
            threads.append(threadstc)
            threadstb = threading.Thread(target=func2,args=(str(queue[1]),))
            threads.append(threadstb)
            threadsta = threading.Thread(target=func1,args=(str(queue[2]),))
            threads.append(threadsta)
        if(i==3):
            threadstd = threading.Thread(target=func4,args=(str(queue[0]),))
            threads.append(threadstd)
            threadstc = threading.Thread(target=func3,args=(str(queue[1]),))
            threads.append(threadstc)
            threadstb = threading.Thread(target=func2,args=(str(queue[2]),))
            threads.append(threadstb)
            threadsta = threading.Thread(target=func1,args=(str(queue[3]),))
            threads.append(threadsta)
        if(i==5):
            threadste = threading.Thread(target=func5,args=(str(queue[0]),))
            threads.append(threadste)
            threadstd = threading.Thread(target=func4,args=(str(queue[1]),))
            threads.append(threadstd)
            threadstc = threading.Thread(target=func3,args=(str(queue[2]),))
            threads.append(threadstc)
            threadstb = threading.Thread(target=func2,args=(str(queue[3]),))
            threads.append(threadstb)
            threadsta = threading.Thread(target=func1,args=(str(queue[4]),))
            threads.append(threadsta)
        if(i==5):
            threadstf = threading.Thread(target=func6,args=(str(queue[0]),))
            threads.append(threadstf)
            threadste = threading.Thread(target=func5,args=(str(queue[1]),))
            threads.append(threadste)
            threadstd = threading.Thread(target=func4,args=(str(queue[2]),))
            threads.append(threadstd)
            threadstc = threading.Thread(target=func3,args=(str(queue[3]),))
            threads.append(threadstc)
            threadstb = threading.Thread(target=func2,args=(str(queue[4]),))
            threads.append(threadstb)
            threadsta = threading.Thread(target=func1,args=(str(queue[5]),))
            threads.append(threadsta)
        if(i==6):
            threadstg = threading.Thread(target=func7,args=(str(queue[0]),))
            threads.append(threadstg)
            threadstf = threading.Thread(target=func6,args=(str(queue[1]),))
            threads.append(threadstf)
            threadste = threading.Thread(target=func5,args=(str(queue[2]),))
            threads.append(threadste)
            threadstd = threading.Thread(target=func4,args=(str(queue[3]),))
            threads.append(threadstd)
            threadstc = threading.Thread(target=func3,args=(str(queue[4]),))
            threads.append(threadstc)
            threadstb = threading.Thread(target=func2,args=(str(queue[5]),))
            threads.append(threadstb)
            threadsta = threading.Thread(target=func1,args=(str(queue[6]),))
            threads.append(threadsta)
        if(i==7):
            threadsth = threading.Thread(target=func8,args=(str(queue[0]),))
            threads.append(threadsth)
            threadstg = threading.Thread(target=func7,args=(str(queue[1]),))
            threads.append(threadstg)
            threadstf = threading.Thread(target=func6,args=(str(queue[2]),))
            threads.append(threadstf)
            threadste = threading.Thread(target=func5,args=(str(queue[3]),))
            threads.append(threadste)
            threadstd = threading.Thread(target=func4,args=(str(queue[4]),))
            threads.append(threadstd)
            threadstc = threading.Thread(target=func3,args=(str(queue[5]),))
            threads.append(threadstc)
            threadstb = threading.Thread(target=func2,args=(str(queue[6]),))
            threads.append(threadstb)
            threadsta = threading.Thread(target=func1,args=(str(queue[7]),))
            threads.append(threadsta)
        if(i==8):
            threadsti = threading.Thread(target=func9,args=(str(queue[0]),))
            threads.append(threadsti)
            threadsth = threading.Thread(target=func8,args=(str(queue[1]),))
            threads.append(threadsth)
            threadstg = threading.Thread(target=func7,args=(str(queue[2]),))
            threads.append(threadstg)
            threadstf = threading.Thread(target=func6,args=(str(queue[3]),))
            threads.append(threadstf)
            threadste = threading.Thread(target=func5,args=(str(queue[4]),))
            threads.append(threadste)
            threadstd = threading.Thread(target=func4,args=(str(queue[5]),))
            threads.append(threadstd)
            threadstc = threading.Thread(target=func3,args=(str(queue[6]),))
            threads.append(threadstc)
            threadstb = threading.Thread(target=func2,args=(str(queue[7]),))
            threads.append(threadstb)
            threadsta = threading.Thread(target=func1,args=(str(queue[8]),))
            threads.append(threadsta)
        if(i==9):
            threadstj = threading.Thread(target=func10,args=(str(queue[0]),))
            threads.append(threadstj)
            threadsti = threading.Thread(target=func9,args=(str(queue[1]),))
            threads.append(threadsti)
            threadsth = threading.Thread(target=func8,args=(str(queue[2]),))
            threads.append(threadsth)
            threadstg = threading.Thread(target=func7,args=(str(queue[3]),))
            threads.append(threadstg)
            threadstf = threading.Thread(target=func6,args=(str(queue[4]),))
            threads.append(threadstf)
            threadste = threading.Thread(target=func5,args=(str(queue[5]),))
            threads.append(threadste)
            threadstd = threading.Thread(target=func4,args=(str(queue[6]),))
            threads.append(threadstd)
            threadstc = threading.Thread(target=func3,args=(str(queue[7]),))
            threads.append(threadstc)
            threadstb = threading.Thread(target=func2,args=(str(queue[8]),))
            threads.append(threadstb)
            threadsta = threading.Thread(target=func1,args=(str(queue[9]),))
            threads.append(threadsta)
        if(i==10):
            threadstk = threading.Thread(target=func11,args=(str(queue[0]),),)
            threads.append(threadstk)
            threadstj = threading.Thread(target=func10,args=(str(queue[1]),),)
            threads.append(threadstj)
            threadsti = threading.Thread(target=func9,args=(str(queue[2]),),)
            threads.append(threadsti)
            threadsth = threading.Thread(target=func8,args=(str(queue[3]),),)
            threads.append(threadsth)
            threadstg = threading.Thread(target=func7,args=(str(queue[4]),),)
            threads.append(threadstg)
            threadstf = threading.Thread(target=func6,args=(str(queue[5]),),)
            threads.append(threadstf)
            threadste = threading.Thread(target=func5,args=(str(queue[6]),),)
            threads.append(threadste)
            threadstd = threading.Thread(target=func4,args=(str(queue[7]),),)
            threads.append(threadstd)
            threadstc = threading.Thread(target=func3,args=(str(queue[8]),),)
            threads.append(threadstc)
            threadstb = threading.Thread(target=func2,args=(str(queue[9]),),)
            threads.append(threadstb)
            threadsta = threading.Thread(target=func1,args=(str(queue[10]),),)
            threads.append(threadsta)
        if(i>10):
            if(queue[len(queue)-1]!=epochs):
                queue.append(queue[len(queue)-1]+1)
            queue.pop(0)
            if(len(queue)>11):
                threadstk = threading.Thread(target=func11,args=(str(queue[0]),),)
                threads.append(threadstk)
                threadstj = threading.Thread(target=func10,args=(str(queue[1]),),)
                threads.append(threadstj)
                threadsti = threading.Thread(target=func9,args=(str(queue[2]),),)
                threads.append(threadsti)
                threadsth = threading.Thread(target=func8,args=(str(queue[3]),),)
                threads.append(threadsth)
                threadstg = threading.Thread(target=func7,args=(str(queue[4]),),)
                threads.append(threadstg)
                threadstf = threading.Thread(target=func6,args=(str(queue[5]),),)
                threads.append(threadstf)
                threadste = threading.Thread(target=func5,args=(str(queue[6]),),)
                threads.append(threadste)
                threadstd = threading.Thread(target=func4,args=(str(queue[7]),),)
                threads.append(threadstd)
                threadstc = threading.Thread(target=func3,args=(str(queue[8]),),)
                threads.append(threadstc)
                threadstb = threading.Thread(target=func2,args=(str(queue[9]),),)
                threads.append(threadstb)
                threadsta = threading.Thread(target=func1,args=(str(queue[10]),),)
                threads.append(threadsta)
            else:
                for l in range(len(queue)):
                    threadstk = threading.Thread(target=funq[l],args=(str(queue[l]),),)
                    threads.append(threadstk)
        for threadnum in range(len(threads)):
            threads[threadnum].start()
        for threadnum in range(len(threads)):
            threads[threadnum].join()
    toc = time.time()
    listparallel.append(toc -tic)
```

```python
listserial = []
for ep in range(11,21):
    tic = time.time()
    for i in range(ep):
        func1(str(i+1))
        func2(str(i+1))
        func3(str(i+1))
        func4(str(i+1))
        func5(str(i+1))
        func6(str(i+1))
        func7(str(i+1))
        func8(str(i+1))
        func9(str(i+1))
        func10(str(i+1))
        func11(str(i+1))    
    toc = time.time()
    listserial.append(toc-tic)
```

```python
xaxis = list(range(11,21))
```

```python
import matplotlib.pyplot as plt
plt.plot(xaxis , listparallel, color = 'tab:blue', label = 'Pipelined')
plt.plot(xaxis, listserial, color = 'tab:green', label = 'Serial')
plt.xlabel("Epochs")
plt.ylabel("Time in seconds")
plt.legend()
plt.show()
```
